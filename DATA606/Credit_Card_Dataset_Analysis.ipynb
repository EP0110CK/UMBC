```python
creditdata = "Credit Card Data.csv"

df = pd.read_csv(creditdata)

print(df.head())
```

```python
print("Columns:", df.columns.tolist())

category_col = "category"   # or 'transaction_category'
amount_col = "amt"       # or 'transaction_amount'

# Basic sanity check/cleaning
print("\nMissing values:\n", df[[category_col, amount_col]].isnull().sum())
print("\nData types:\n", df[[category_col, amount_col]].dtypes)

# Group by category to get insights
category_summary = df.groupby(category_col)[amount_col].agg(['count', 'sum', 'mean', 'median', 'max', 'min']).reset_index()
category_summary = category_summary.sort_values(by='sum', ascending=False)

print("\nðŸ“Š Transaction Summary by Category:\n", category_summary)

# Bar plot: Total amount per category
plt.figure(figsize=(12, 6))
sns.barplot(data=category_summary, x=category_col, y='sum', palette='viridis')
plt.xticks(rotation=45)
plt.title("Total Transaction Amount by Category")
plt.ylabel("Total Amount")
plt.xlabel("Transaction Category")
plt.tight_layout()
plt.show()

# Boxplot: Distribution of transaction amount per category
plt.figure(figsize=(14, 6))
sns.boxplot(data=df, x=category_col, y=amount_col)
plt.xticks(rotation=45)
plt.title("Transaction Amount Distribution by Category")
plt.tight_layout()
plt.show()
```

```python
date_col = "trans_date_trans_time"        # or 'transaction_date'
amount_col = "amt"    # or 'transaction_amount'
category_col = "category"

df[date_col] = pd.to_datetime(df[date_col], errors='coerce')

print("Missing dates:", df[date_col].isna().sum())

df['year'] = df[date_col].dt.year
df['month'] = df[date_col].dt.month
df['day'] = df[date_col].dt.day
df['weekday'] = df[date_col].dt.day_name()
df['hour'] = df[date_col].dt.hour

# 1. Total transaction amount per month
monthly_trend = df.groupby([df[date_col].dt.to_period('M')])[amount_col].sum().reset_index()
monthly_trend[date_col] = monthly_trend[date_col].dt.to_timestamp()

plt.figure(figsize=(12, 5))
sns.lineplot(data=monthly_trend, x=date_col, y=amount_col, marker='o')
plt.title("Total Transaction Amount per Month")
plt.xlabel("Month")
plt.ylabel("Total Amount")
plt.grid(True)
plt.tight_layout()
plt.show()

# 2. Average transaction amount by weekday
weekday_avg = df.groupby('weekday')[amount_col].mean().reindex([
    'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'
])

plt.figure(figsize=(10, 4))
sns.barplot(x=weekday_avg.index, y=weekday_avg.values, palette='pastel')
plt.title("Average Transaction Amount by Weekday")
plt.ylabel("Average Amount")
plt.xlabel("Day of the Week")
plt.tight_layout()
plt.show()
```

```python
df = df.dropna(subset=[date_col, category_col, amount_col])

# UI function to navigate each category trend
def plot_category_trend(selected_category):
    filtered_df = df[df[category_col] == selected_category]

    monthly_trend = (
        filtered_df
        .groupby(filtered_df[date_col].dt.to_period('M'))[amount_col]
        .sum()
        .reset_index()
    )
    monthly_trend[date_col] = monthly_trend[date_col].dt.to_timestamp()

    plt.figure(figsize=(12, 5))
    sns.lineplot(data=monthly_trend, x=date_col, y=amount_col, marker='o')
    plt.title(f"Monthly Transaction Trend for Category: {selected_category}")
    plt.ylabel("Total Transaction Amount")
    plt.xlabel("Month")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Dropdown menu
category_dropdown = widgets.Dropdown(
    options=sorted(df[category_col].dropna().unique()),
    description='Category:',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='50%')
)

interact(plot_category_trend, selected_category=category_dropdown)
```
